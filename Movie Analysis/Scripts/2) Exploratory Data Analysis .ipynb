### 1) Importing Libraries and Dataframe

```python
# Importing libraries

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Importing dataframe

path = r'/Users/matthewabrams/Desktop/Movie Analysis'
movies = pd.read_csv(os.path.join(path, 'Data', 'Prepared Data', 'movies_modified.csv'), index_col = False)
movies.drop(columns = ['Unnamed: 0'], inplace = True)
movies
```

### 2) Exploratory Data Analysis

#### 2a) Graphing values related to Country

```python
# Bar graph involving count of films produced by country
# Majority of films are released in the US

movies['country'].value_counts().nlargest(10).sort_values().plot.barh(figsize = (7,7), color = 'red')
plt.title('Count of Films from 1980 to 2020 by Country (Top 10 Countries)', fontweight = 'bold', fontsize = 12)
plt.xlabel('Count of Films Released', fontweight = 'bold', fontsize = 12)
plt.ylabel('Country', fontweight = 'bold', fontsize = 12)
plt.xticks(fontsize = 11)
plt.yticks(fontsize = 11)
```

```python
# Bar graph involving average profit per film grouped by country
# Even though more films are released in US, some movie companies make higher profit per film on average in China
# Movie companies have earned profit releasing films in China, US, Lebanon, Cameroon and South Korea 
# Movie companies that released films in all other countries ended up with net losses in the long run, their budget exceeds their revenue

movies.groupby('country')['profit'].mean().sort_values().plot.barh(figsize = (7,7), color = 'red')
plt.title('Average Profit Per Film Made from 1980 to 2020', fontweight = 'bold', fontsize = 12)
plt.xlabel('Profit', fontweight = 'bold', fontsize = 12)
plt.ylabel('Country', fontweight = 'bold', fontsize = 12)
plt.xticks(fontsize = 11)
plt.yticks(fontsize = 11)
```

```python
# Bar graph involving average profit per film grouped by country (Top 5 Countries by Highest Profit Earnings)

movies.groupby('country')['profit'].mean().nlargest(5).sort_values().plot.barh(figsize = (7,7), color = 'red')
plt.title('Average Profit Per Film Released from 1980 to 2020 (Top 5 Countries)', fontweight = 'bold', fontsize = 12)
plt.xlabel('Profit', fontweight = 'bold', fontsize = 12)
plt.ylabel('Country', fontweight = 'bold', fontsize = 12)
plt.xticks(fontsize = 11)
plt.yticks(fontsize = 11)
```

```python
# Bar graph involving average gross per film grouped by country
# Movie companies make more gross earnings on average per film released in China followed by US

movies.groupby('country')['gross'].mean().nlargest(10).sort_values().plot.barh(figsize = (7,7), color = 'red')
plt.title('Average Gross Per Film Released from 1980 to 2020 (Top 10 Countries)', fontweight = 'bold', fontsize = 12)
plt.xlabel('Gross', fontweight = 'bold', fontsize = 12)
plt.ylabel('Country', fontweight = 'bold', fontsize = 12)
plt.xticks(fontsize = 11)
plt.yticks(fontsize = 11)
```

#### 2c) Graphing values related to genre

```python
# Bar graph of average gross figure per film by genre
# Movie companies make more gross figures in releasing Animation, Family, and Action films 

movies.groupby('genre')['gross'].mean().sort_values().plot.barh(figsize = (7,7), color = 'blue')
plt.title('Average Gross Earnings for Genre Films 1980 to 2020', fontweight = 'bold', fontsize = 12)
plt.xlabel('Gross', fontweight = 'bold', fontsize = 12)
plt.ylabel('Genre', fontweight = 'bold', fontsize = 12)
plt.xticks(fontsize = 11)
plt.yticks(fontsize = 11)
```

```python
# Bar graph of average profit earned per film by genre
# Movie companies make more profit in Animation, Family and Action films 
# Movie companies end with net losses on average releasing Western, Musicals/Music, and Sports, their budget exceeds their gross figures

movies.groupby('genre')['profit'].mean().sort_values().plot.barh(figsize = (7,7), color = 'blue')
plt.title('Average Profit Earnings for Genre Films 1980 to 2020', fontweight = 'bold', fontsize = 12)
plt.xlabel('Gross', fontweight = 'bold', fontsize = 12)
plt.ylabel('Genre', fontweight = 'bold', fontsize = 12)
plt.xticks(fontsize = 11)
plt.yticks(fontsize = 11)
```

```python
top_genre = movies[movies['genre'].isin(['Animation', 'Family', 'Action', 'Mystery', 'Adventure'])]
top_genre
```

```python
top_genre.groupby('star')['gross'].sum().sort_values(ascending = False).head(20).plot.barh(figsize = (7,7), color = 'orange')
plt.title('Top 10 Stars Based on Top 5 Genres By Total Gross Earnings (1980-2020)', fontweight = 'bold', fontsize = 10)
plt.xlabel('Gross', fontweight = 'bold', fontsize = 12)
plt.ylabel('Star', fontweight = 'bold', fontsize = 12)
plt.xticks(fontsize = 11)
plt.yticks(fontsize = 11)
```

### 3. Finding relationships between numerical variables

```python
# Creating a subset dataframe for numerical variables labeled sub

sub = movies[['IMDb score', 'votes', 'budget', 'gross', 'profit', 'duration (in min)']]
```

```python
# Creating a correlation matrix for sub dataframe, rounding r values to two decimal places
# Budget and IMDb user votes are more correlated with gross figures

sub.corr().round(2)
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(sub.corr(), fignum = f.number) # type of plot
plt.xticks(range(sub.shape[1]), sub.columns,  fontsize=10, rotation = 45) # x axis labels
plt.yticks(range(sub.shape[1]), sub.columns,  fontsize=10) # y axis labels
cb = plt.colorbar() # add a color legend (called colorbar)
cb.ax.tick_params(labelsize = 10) # add font size
plt.title('Correlation Matrix', fontsize = 14) # add title
plt.show()
```

```python
f,ax = plt.subplots(figsize=(10,10))
```

```python
# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.

corr_2 = sns.heatmap(sub.corr(), annot = True) 
plt.xticks(rotation = 45)

# The `annot` argument allows the plot to place the correlation coefficients onto the heatmap.
```

### 4) Scatter Plots

```python
# Created scatter plot between budget and gross columns
# There is a strong positive relationship between these two columns by a coefficient of 0.75
# This indicates that a high budget equates to a higher gross figure for a film

sns.lmplot(x = 'budget', y = 'gross', data = sub)
```

```python
# Created scatter plot between budget and gross columns
# There is a medium to strong positive relationship between these two columns by a coefficient of 0.63

sns.lmplot(x = 'votes', y = 'gross', data = sub)
```

```python
# Created scatter plot between profit and gross columns
# There is an almost perfect positive relationship between these two columns by a coefficient of 0.99

sns.lmplot(x = 'gross', y = 'profit', data = sub)
```

```python
# Created scatter plot between score and profit columns
# There is a weak relationship between these two columns by a coefficient of 0.2, seems that IMDb score has little impact on profit

sns.lmplot(x = 'IMDb score', y = 'profit', data = sub)
```

### 5) Pair Plots

```python
# Created a pairplot for sub dataframe

sns.pairplot(sub)
```

```python
# It shows a the distribution of all numeric variables while also creating scatterplots for every variable crossed with every other variable 
# I want to explore more information between budget and gross along with votes and gross since these variables have a closer relationhip with one another than other variables
# The strongest positive correlation we can see is between 'gross' and 'profit'. This would make sense - generally the films with the highest revenues will have the highest profits, and the films with the lowest revenues will have the lowest profits.
# Similarly, 'budget' and 'gross' have a strong correlation - again, this makes sense. The more money that is invested in a movie, the more money you'd expect to make
# However there's no strong correlation between 'budget' and actual 'profit', suggesting that many films in this data set performed worse at box office than would be expected based on their budget.
```

### 6) Categorical Plots

```python
# Use a histogram to visualize the distribution of the variables. 

sns.distplot(movies['gross'], bins = 50)
```

```python
# Derived a new variable based off of gross column. 
# If gross figure less than 20000000, then < 20 M
# If gross figure between 20000000 and 100000000, then '0-100 M
# If gross figure more than 100000000, then > 100 M

movies.loc[movies['gross'] < 20000000, 'gross category'] = '< 20 M'
movies.loc[(movies['gross'] >= 20000000) & (movies['gross'] <= 100000000), 'gross category'] = '20-100 M'
movies.loc[movies['gross'] > 100000000, 'gross category'] = '> 100 M'
```

```python
movies['gross category'].value_counts()
```

```python
# Created a categorical plot in seaborn using the gross categories created above
# Relationship between number of IMDb user votes and gross figure
# All types of values by gross category are on top of each other
# I've decided to plot seperate scatter plots based off of gross category type

sns.set(style = "ticks")
g = sns.scatterplot(x = "votes", y = "gross", hue = "gross category", data = movies)
```

```python
# Created a subset of movies dataframe where gross category is relatively low, named this subset low_gross

low_gross = movies[movies['gross category']=='< 20 M']
low_gross
```

```python
# Graphed low_gross datafame

sns.scatterplot(x = "votes", y = "gross", data = low_gross, color = 'green')
```

```python
# Created a subset of movies dataframe where gross category is the typical amount, named this subset middle_gross

middle_gross = movies[movies['gross category']=='20-100 M']
middle_gross
```

```python
# Graphed middle_gross datafame

sns.scatterplot(x = "votes", y = "gross", data = middle_gross)
```

```python
# Created a subset of movies dataframe where gross category is the typical amount, named this subset high_gross

high_gross = movies[movies['gross category']=='> 100 M']
high_gross
```
